```python
# TUGAS 1

from pymongo import MongoClient

# Hubungkan ke MongoDB lokal
client = MongoClient("mongodb://localhost:27017/")

# Pilih database dan koleksi
db = client["company_db"]
collection = db["employees"]

print("connected to MongoDB")
```

```python
# TUGAS 2: Insert beberapa data tambahan, lakukan update, serta delete data

# Contoh Insert Data
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}
collection.insert_one(employee_data)
print('Data inserted')
```

```python
# tugas 2
# Insert beberapa data tambahan
employees = [
    {'name': 'Bob', 'department': 'Finance', 'age': 35, 'salary': 5200},
    {'name': 'Charlie', 'department': 'IT', 'age': 27, 'salary': 6000},
    {'name': 'Daisy', 'department': 'IT', 'age': 23, 'salary': 4800},
    {'name': 'Eve', 'department': 'HR', 'age': 31, 'salary': 5100},
    {'name': 'Frank', 'department': 'HR', 'age': 26, 'salary': 4700},
]
collection.insert_many(employees)
print("Data tambahan berhasil dimasukkan.")

# Contoh Query Data
for employee in collection.find():
    print(employee)
```

```python
# tugas 2

# Update gaji karyawan bernama Daisy
collection.update_one(
    {'name': 'Daisy'}, 
    {'$set': {'salary': 5500}}
)
print("✅ Data Daisy berhasil diperbarui.")
print("Hasil update:")
print(collection.find_one({'name': 'Daisy'}, {'_id': 0, 'name': 1, 'salary': 1}))
```

```python
# Hapus data karyawan bernama Frank
deleted = collection.delete_one({'name': 'Frank'})
if deleted.deleted_count > 0:
    print("Data Frank berhasil dihapus.")
else:
    print("Data Frank tidak ditemukan.")

# Tampilkan data sisa di departemen HR
print("Data HR setelah penghapusan:")
for emp in collection.find({'department': 'HR'}, {'_id': 0, 'name': 1, 'salary': 1}):
    print(emp)
```

```python
# tugas 2

# lihat data karyawan sekarang
print("Data karyawan saat ini:")
for employee in collection.find({}, {'_id': 0, 'name': 1, 'department': 1, 'salary': 1}):
    print(employee)
```

```python
# tugas 3

# Query Aggregation untuk mencari rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
# Latihan 4.2: Lakukan query untuk menemukan produk dengan harga di atas rata-rata.

# Hitung rata-rata harga
avg_price = list(collection.aggregate([
    {'$group': {'_id': None, 'average_price': {'$avg': '$price'}}}
]))[0]['average_price']
print(f"Rata-rata harga produk: {avg_price:.2f}\n")

# Tampilkan produk dengan harga di atas rata-rata
result = collection.find({'price': {'$gt': avg_price}})
print("Produk dengan harga di atas rata-rata:\n")
for data in result:
    print(data)
```

```python
# Latihan 4.3: Buatlah aggregation pipeline untuk menghitung total produk dalam setiap kategori.

pipeline = [
    {'$group': {'_id': '$category', 'total_products': {'$sum': 1}}}
]
result = collection.aggregate(pipeline)
print("Total produk dalam setiap kategori:\n")
for data in result:
    print(data)
```

```python
# Tugas 1: Cari 5 karyawan dengan gaji tertinggi dalam setiap departemen, gunakan query atau aggregation yang sesuai.

print("5 Karyawan dengan Gaji Tertinggi per Departemen:")

pipeline_1 = [
    {"$sort": {"department": 1, "salary": -1}},
    {"$group": {
        "_id": "$department",
        "top_employees": {"$push": {"name": "$name", "salary": "$salary"}}
    }},
    {"$project": {
        "top_5": {"$slice": ["$top_employees", 5]}
    }}
]

for doc in collection.aggregate(pipeline_1):
    print(f"\nDepartemen: {doc['_id']}")
    for emp in doc['top_5']:
        print(f" Name: {emp['name']:10} | Salary: {emp['salary']:,}")
print()
```

```python
# Tugas 2: Buatlah skenario di mana Anda harus menghapus karyawan yang berusia di bawah 25 tahun dari database.

print("=== Menghapus Karyawan Berusia di Bawah 25 Tahun ===")
deleted = collection.delete_many({"age": {"$lt": 25}})
print(f"Jumlah data yang dihapus: {deleted.deleted_count}\n")
```

```python
# Tugas 3: Buatlah laporan ringkas (menggunakan MongoDB query) yang menghitung total gaji karyawan di setiap departemen, serta rata-rata umur karyawan.

print("Total Gaji dan Rata-rata Umur per Departemen\n")

pipeline_3 = [
    {"$group": {
        "_id": "$department",
        "total_salary": {"$sum": "$salary"},
        "average_age": {"$avg": "$age"}
    }}
]

for doc in collection.aggregate(pipeline_3):
    print(f"Departemen: {doc['_id']}")
    print(f"Total Gaji   : Rp{doc['total_salary']:,}")
    print(f"Rata-rata Umur: {doc['average_age']:.1f} tahun\n")
```
